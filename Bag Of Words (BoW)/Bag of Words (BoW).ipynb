# Bag Of Words (BoW)

- The Bag of Words (BoW) model is one of the most fundamental and classical text representation techniques in Natural Language Processing (NLP).
- It converts text data into numerical vectors so that machine learning algorithms can understand and process text data.
- Bag of Words is a technique that represents a document as a collection of word counts without considering grammar or word order.
- Bag of Words is a text representation technique that converts text into numerical vectors based on word frequency, ignoring grammar and word order.

## Code Implementation

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
"I love NLP",
"NLP is amazing",
"I love machine learning"
]

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(documents)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("BoW Matrix:\n", X.toarray())
```
