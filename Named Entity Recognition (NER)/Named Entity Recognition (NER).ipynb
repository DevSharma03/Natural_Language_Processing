# Named Entity Recognition (NER)

- Named Entity Recognition (NER) is a Natural Language Processing (NLP) technique used to identify and classify important information (entities) in text into predefined categories.
- Named Entity Recognition (NER) is an NLP task that identifies and classifies entities such as persons, organizations, locations, dates, and other real-world objects in text.

These entities usually include:

1. Person names
2. Organizations
3. Locations
4. Dates and time
5. Money
6. Percentages
7. Products
8. Events

NER helps convert unstructured text → structured information.

### What is an Entity?
- An entity is a real-world object that can be uniquely identified.

Example sentence:
```
"Elon Musk founded SpaceX in 2002 in the United States."
```
NER will detect:
| Word          | Entity Type  |
| ------------- | ------------ |
| Elon Musk     | PERSON       |
| SpaceX        | ORGANIZATION |
| 2002          | DATE         |
| United States | LOCATION     |

```
Output Representation:
[Elon Musk] -> PERSON
[SpaceX] -> ORGANIZATION
[2002] -> DATE
[United States] -> LOCATION
```

### Types of Named Entities

Common Entity Categories
| Entity      | Meaning              | Example      |
| ----------- | -------------------- | ------------ |
| PERSON      | Name of a person     | Elon Musk    |
| ORG         | Organization         | Google       |
| GPE         | Geo-political entity | India        |
| LOC         | Location             | Himalayas    |
| DATE        | Date                 | 12 January   |
| TIME        | Time                 | 10 AM        |
| MONEY       | Monetary value       | $500         |
| PERCENT     | Percentage           | 20%          |
| PRODUCT     | Product              | iPhone       |
| EVENT       | Event                | Olympics     |
| LAW         | Legal document       | Constitution |
| WORK_OF_ART | Books/movies         | Titanic      |

## Code Example

````python
# Using SpaCy

import spacy

nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded SpaceX in California in 2002."

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "-->", ent.label_)
````
