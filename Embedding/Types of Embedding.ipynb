# Summary 

| Type                  | What it Represents         | Example         |
| --------------------- | -------------------------- | --------------- |
| Word Embeddings       | Individual words           | Word2Vec, GloVe |
| Character Embeddings  | Characters                 | FastText        |
| Subword Embeddings    | Word pieces                | BERT tokenizer  |
| Contextual Embeddings | Meaning depends on context | BERT, GPT       |
| Sentence Embeddings   | Whole sentence             | Sentence-BERT   |
| Document Embeddings   | Entire document            | Doc2Vec         |
| Token Embeddings      | Tokens in transformers     | BERT/GPT        |
| Positional Embeddings | Word order                 | Transformers    |

## Types of Embeddings Used in AI Systems

| System                 | Embedding Type                           |
| ---------------------- | ---------------------------------------- |
| LLMs (GPT, LLaMA)      | Token Embeddings + Positional Embeddings |
| RAG systems            | Sentence/Text Embeddings                 |
| Semantic search        | Sentence Embeddings                      |
| Recommendation systems | User/item embeddings                     |
| Vector databases       | Text embeddings                          |
