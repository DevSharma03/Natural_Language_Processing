# Embedding

- An embedding is an dence numerical vector representation of data.
- Embedding is a dense vector representation of words or data that captures semantic relationships in a continuous vector space.
- Embedding is a technique that converts words into numerical vectors so that machines can understand the meaning and relationships between words.
- Embedding = converting words into meaningful numbers (vectors). As computers cannot understand text like humans, so we convert words into lists of numbers that represent their meaning.




---

## Embedding Space
- Embeddings exist in a vector space.
- Words with similar meanings are closer.
